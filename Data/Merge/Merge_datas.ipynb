Leeremos y crearemos los datasets


```python
import pandas as pd

df_covid = pd.read_csv('covid-19_datos_por_comuna.csv')
df_socio = pd.read_csv('casen_2017_2020_2022_unido.csv')
```

```python
df_covid.head()
```

```python
comunas_covid = df_covid['Comuna'].unique()
print(sorted(comunas_covid))
```

```python
comunes = set(df_socio['Comuna'].unique()) & set(df_covid['Comuna'].unique())
print(comunes)
```

```python
df_merged_inner = pd.merge(df_socio, df_covid, on=['Comuna', 'Año'], how='inner')
```
