Primero estandarizaremos los dataframes

```python
import pandas as pd

df_2017 = pd.read_csv("CASEN_POBREZA_GEOGRAFICA_2017.csv")
df_2020 = pd.read_csv("CASEN_PANDEMIA_POBREZA_GEOGRAFICA_2020.csv")
df_2022 = pd.read_csv("CASEN_POBREZA_GEOGRAFICA_2022.csv") 
```

```python
# Para CASEN 2017 - Renombrar y seleccionar columnas comunes
df_2017_clean = df_2017.rename(columns={
    'Pobreza': 'Pobreza Categoria',
    'Pobreza Multidimensional': 'Pobreza_multidimensional', 
    'Año': 'Año',
    'Sistema Provisional': 'Sistema_previsional',
    'Nivel socioeconomico': 'NSE',
    'Ingreso por Hogar': 'Ingreso_hogar',
    'Sexo': 'Sexo'
})[['folio', 'id_vivienda', 'Pobreza Categoria', 'Pobreza_multidimensional', 'Año', 
     'Sistema_previsional', 'NSE', 'region', 'edad', 'Sexo', 'Ingreso_hogar', 'Comuna']]

df_2017_clean['id_persona'] = 1 
```

```python
# Para CASEN 2020 - Estandarizar
df_2020_clean = df_2020.rename(columns={
    'Pobreza': 'Pobreza Categoria',
    'Año': 'Año',
    'Sistema Provisional': 'Sistema_previsional', 
    'Nivel socioeconomico': 'NSE',
    'Ingreso por Hogar': 'Ingreso_hogar',
    'Sexo': 'Sexo'
})[['folio', 'id_persona', 'Pobreza Categoria', 'NSE', 'pobreza_sinte', 'region', 
     'Año', 'Sistema_previsional', 'edad', 'Sexo', 'Ingreso_hogar', 'Comuna']]

df_2020_clean['id_vivienda'] = df_2020_clean['folio']
```

```python
# Para CASEN 2022 - Estandarizar  
df_2022_clean = df_2022.rename(columns={
    'Pobreza': 'Pobreza Categoria',
    'Pobreza Multidimensional': 'Pobreza_multidimensional',
    'Año': 'Año',
    'Sistema Provisional': 'Sistema_previsional',
    'Nivel socioeconomico': 'NSE',
    'Ingreso por Hogar': 'Ingreso_hogar', 
    'Sexo': 'Sexo'
})[['folio', 'id_persona', 'Año', 'Pobreza Categoria', 'Pobreza_multidimensional',
     'Sistema_previsional', 'NSE', 'region', 'edad', 'Sexo', 'Ingreso_hogar', 
      'Comuna']]

df_2022_clean['id_vivienda'] = df_2022_clean['folio']
```

Unir los DataFrames

```python
df_completo = pd.concat([df_2017_clean, df_2020_clean, df_2022_clean], 
                       ignore_index=True, 
                       sort=False)
```

```python
df_completo['pobreza_sinte'] = df_completo['pobreza_sinte'].fillna('No aplica')
```
